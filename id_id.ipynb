```python
#####test#####

def def_identity(df):
    # create empty dictionary to store identities
    identity_dict = {}
    output_list = []
    output_dict = {'案件':'', '被告':{}}
    
    # iterate through each case in the dataframe
    for case in df['案件'].unique():
        # create empty list to store people with unknown identity
        unknown_people = []
        
        # iterate through each row in the dataframe for the current case
        for _, row in df[df['案件'] == case].iterrows():
            # check relationship type and update identity_dict accordingly
            if row['关系'] == '同主犯':
                identity_dict[row['人物1']] = '主犯'
                identity_dict[row['人物2']] = '主犯'
            elif row['关系'] == '同从犯':
                identity_dict[row['人物1']] = '从犯'
                identity_dict[row['人物2']] = '从犯'
            else:
                # add people to list of unknown identity
                unknown_people.append(row['人物1'])
                unknown_people.append(row['人物2'])
        
        # remove duplicates from unknown_people list
        unknown_people = list(set(unknown_people))
        
        # generate possibility lists for unknown people
        possibility_lists = []
        for person in unknown_people:
            possibilities = ['主犯', '从犯']
            for _, row in df[df['案件'] == case].iterrows():
                if person == row['人物1'] and row['人物2'] in identity_dict:
                    if row['关系'] == '主从犯':
                        if identity_dict[row['人物2']] == '主犯':
                            possibilities = ['从犯']
                        else:
                            possibilities = ['主犯']
                        break
                elif person == row['人物2'] and row['人物1'] in identity_dict:
                    if row['关系'] == '主从犯':
                        if identity_dict[row['人物1']] == '主犯':
                            possibilities = ['从犯']
                        else:
                            possibilities = ['主犯']
                        break
            possibility_lists.append((person, possibilities))
        
        # remove possibilities that conflict with identities in identity_dict
        for person, possibilities in possibility_lists:
            for identity_person, identity_gender in identity_dict.items():
                if person != identity_person and person in possibility_lists:
                    if identity_gender in possibilities:
                        possibilities.remove(identity_gender)
            identity_dict[person] = possibilities[0]
        
        # # add identity_dict to final result dictionary
        # identity_dict_copy = identity_dict.copy()
        # # identity_dict.clear()
        # identity_dict.update({case: identity_dict_copy})
        print(case)
        print(identity_dict)
        

        output_list.append(output_dict)
        identity_dict.clear()
    
    return output_list
```
